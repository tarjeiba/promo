**Oppgave 8** Lag en funksjon `tetrahedertall(n)` som tar inn et heltall `n` og gir tilbake det $n$-te tetrahedertallet.

En *rekursiv funksjon* er en funksjon hvor funksjonen selv blir kalt i sin egen definisjon. Et rart eksempel vises under, hvor funksjonen `tell_ned_fra` tar inn et heltall `n` og skriver ut verdiene fra og med $n$ til og med $1$.

```python
def tell_ned_fra(n):
    print(n)
    if n == 1:
        return None
    else:
        tell_ned_fra(n - 1)  # her gj√∏res det rekursivet kallet

tell_ned_fra(5)
```

Her skulle funksjonen bare skrive ut noe, og vi var ikke interessert i funksjonens "returverdi". Oftere vil det v√¶re at vi er ute etter returverdien til funksjonen (f.eks. fordi den skal tildeles et variabelnavn), og da m√• vi holde styring p√• at det er en reell returverdi i alle niv√•ene av det rekursive kallet.

Funksjonen `rekursiv_multiplikasjon` som defineres under, tar inn to positive heltall, `a` og `b`, og returnerer produktet deres.

```python
def rekursiv_multiplikasjon(a, b):
    if b == 1:
        return a
    else:
        return a + rekursiv_multiplikasjon(a, b-1)
    
rekursiv_multiplikasjon(13, 15)
```

Hvor det kan v√¶re n√∏dvendig √• spesifisere et par elementer:
1. "the infinite sum" betyr en sum hvor vi har med uendelig mange ledd, $\sum_{n=0}^\infty n = 0 + 1 + 2 + \cdots$.

1. "telescoping series" er en rekke hvor vi kan bli kvitt summasjonstegnet ved √• skyfle om p√• termene.
